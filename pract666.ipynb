```python
skyrim = pd.read_csv('Skyrim_Weapons.csv')
plt.bar(skyrim['Name'].head(4),skyrim['Weight'].head(4), label = 'Вес', color = 'red', alpha = 0.5)
plt.legend()
plt.show()
```

Это столбчатая диаграмма соотношения стоимости, касательно названия. Сначала читается файл, затем данные выбираются по первым четырем названиям в файле, столбики диаграммы означают вес, сами они голубые и, благодаря параметру "alpha" полупрозрачные. Маркеры здесь не нужны, т.к. это диаграмма. Снизу подписано название оружия, а слева - значения его веса. plt.legend() показывает легенду, а plt.show() саму диаграмму

```python
skyrim = pd.read_csv('Skyrim_Weapons.csv')
plt.bar(skyrim['Damage'].head(4),skyrim['Gold'].head(4), label = 'Урон', color = 'purple')
plt.bar(skyrim['Weight'].head(4),skyrim['Gold'].head(4), label = 'Вес', color = 'green', alpha = 0.5)
plt.legend()
plt.show()
```

Это столбчатая диаграмма соотношения стоимости, касательно урона и веса. Сначала читается файл, затем данные выбираются по первым четырем названиям в файле, столбики диаграммы означают вес и урон, сами они фиолетовые и голубые, которые, благодаря параметру "alpha", полупрозрачные. Маркеры здесь не нужны, т.к. это диаграмма. Снизу подписаны значения урона и веса, а слева - значения его стоимости. plt.legend() показывает легенду, а plt.show() саму диаграмму

```python
skyrim = pd.read_csv('Skyrim_Weapons.csv')
data = skyrim['Damage'].value_counts().head(6)
plt.pie(data,labels=data.index,autopct='%1.1f%%', colors = ['yellow','red'])
plt.legend()
plt.show()
```

Это круговая диаграмма, которая содержит в себе процентное соотношение урона оружий. Сначала читается файл, затем данные выбираются по первым шести названиям в файле, заполняется список информации с помощью value_counts(). Autopct='%1.1f%%' значит, что данные будут представлены в процентах. Названия будут подписаны благодаря коллекции.

```python
skyrim = pd.read_csv('Skyrim_Weapons.csv')
data = skyrim['Gold'].value_counts().tail(8)
plt.pie(data,labels=data.index,autopct='%1.1f%%', colors = ['brown','blue'])
plt.legend()
plt.show()
```

Это круговая диаграмма, которая содержит в себе процентное соотношение стоимости оружия. Сначала читается файл, затем данные выбираются по последним восьми названиям в файле, заполняется список информации с помощью value_counts(). Autopct='%1.1f%%' значит, что данные будут представлены в процентах. Названия будут подписаны благодаря коллекции.
